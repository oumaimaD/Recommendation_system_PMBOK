DATA CLEANING

```python
def remove_irr_char_func(text):
    return re.sub(r'[^a-zA-Z]', ' ', text)
```

```python
def remove_dup(text):
    return TextBlob(text).correct()
```

```python
def wordlowercase(text):
    return text.lower()
```

```python
def remove_punctuation(text):
    return re.sub(r'[^a-zA-Z0-9]', ' ', text)
```

```python
def remove_whitespaces(text):
    return re.sub(r'^\s*|\s\s*', ' ', text).strip()
```

```python
def remove_nonascii(text):
    return re.sub(r'[^\x00-\x7f]',r'', text) # or ''.join([x for x in text if x in string.printable]) 
```

```python
def remove_html(text):
    html = re.compile(r"<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});")
    return re.sub(html, "", text)
```

```python
def remove_URL(text):
    return re.sub(r"https?://\S+|www\.\S+", "", text)
```

```python
df['Cleantexte'] = df['texte'].str.lower()
df['Cleantexte'] = df['Cleantexte'].apply(remove_irr_char_func)
df['Cleantexte'] = df['Cleantexte'].apply(wordlowercase)
df['Cleantexte'] = df['Cleantexte'].apply(remove_punctuation)
df['Cleantexte'] = df['Cleantexte'].apply(remove_whitespaces)
df['Cleantexte'] = df['Cleantexte'].apply(remove_nonascii)
df['Cleantexte'] = df['Cleantexte'].apply(remove_URL)
df['Cleantexte'] = df['Cleantexte'].apply(remove_html)
```
